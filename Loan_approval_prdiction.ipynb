```python
#handling missing value

df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
#convert null value in to datatype
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mean())

```

```python
df['Total_applicantincome']= df['ApplicantIncome'] + df['CoapplicantIncome']
df.drop(columns=['ApplicantIncome','CoapplicantIncome'],axis=1,inplace=True)
df.head()
```

```python
#convet string value into int
#using label encoding
from sklearn.preprocessing import LabelEncoder
cols = ['Gender','Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status']
label =  LabelEncoder()

for col in cols:
  df[col]=label.fit_transform(df[col])

df.head()
```

```python
#split train data and test data
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

normalization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
tree=DecisionTreeClassifier()
random=RandomForestClassifier()
logis=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=7)
lst=[tree,random,logis,knn]
```

```python
from sklearn.metrics import accuracy_score,classification_report
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('accuracy_score',accuracy_score(y_test,y_pred))
  print('*'*10)
  print('classification reporrt',classification_report(y_test,y_pred))
```
